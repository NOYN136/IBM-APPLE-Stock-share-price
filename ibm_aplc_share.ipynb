```python
import requests
import pandas_datareader.data as web
from datetime import datetime,timedelta
import pandas as pd
import json
import mysql.connector
import yfinance as yf
# replace the "demo" apikey below with your own key from https://www.alphavantage.co/support/#api-key
url = 'https://www.alphavantage.co/query?function=TIME_SERIES_DAILY&symbol=IBM&apikey=demo'
r = requests.get(url)
ibm_data = r.json()
```

```python
print(ibm_data)
```

```python
end = datetime.today()
back_week = 3
ibm_share_price_data = pd.DataFrame()
i=0;j=0
while i < (back_week*7):
    date = (end - timedelta(days =j)).strftime('%Y-%m-%d')
    df = ibm_data['Time Series (Daily)'].get(date,None)
    if df !=None:
        process_data = {"symbol":"IBM","date":date,"open_price":df['1. open'],"close_price":df['4. close'],"volume":df['5. volume']}
        print(process_data)
        process_df = pd.DataFrame([process_data])
        ibm_share_price_data = pd.concat([ibm_share_price_data,process_df])
        i+=1
    j+=1
```

```python
ibm_share_price_data = ibm_share_price_data.reset_index(drop=True)
ibm_share_price_data[['open_price','close_price']] = ibm_share_price_data[['open_price','close_price']].astype(float)
```

```python
print(ibm_share_price_data)
```

```python
apple = yf.Ticker("AAPL")
```

```python
apple_share_price_data = apple.history(period=str(back_week*7)+'d')
apple_share_price_data.reset_index(inplace=True)
apple_share_price_data['Date'] = apple_share_price_data['Date'].apply(lambda x: x.strftime('%Y-%m-%d'))
apple_share_price_data['symbol'] = 'AAPL'
apple_share_price_data = apple_share_price_data[['symbol','Date','Open','Close','Volume']]
apple_share_price_data.rename(columns={'Date':'date','Open':'open_price','Close':'close_price','Volume':'volume'},inplace = True)
apple_share_price_data[['open_price','close_price']] = apple_share_price_data[['open_price','close_price']].astype(float)
print(apple_share_price_data)
```

```python
financial_data = pd.concat([ibm_share_price_data,apple_share_price_data])
financial_data = financial_data.reset_index(drop=True)
print(financial_data)
```
